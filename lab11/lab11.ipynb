```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

# --------------------------------------------------
# Data points from question 2
# --------------------------------------------------
points = {
    'P1': (1, 1),
    'P2': (3, 2),
    'P3': (9, 1),
    'P4': (3, 7),
    'P5': (7, 2),
    'P6': (9, 7),
    'P7': (4, 8),
    'P8': (8, 3),
    'P9': (1,4)
}


labels = list(points.keys())
data = np.array(list(points.values()))

# --------------------------------------------------
# (a) Plot scatter graph
# --------------------------------------------------
for i, label in enumerate(labels):
    plt.scatter(data[i, 0], data[i, 1], c='red')
    plt.text(data[i, 0]+0.1, data[i, 1]+0.1, label)
plt.title("Scatter plot of data points")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# --------------------------------------------------
# (b) Display proximity matrices
# --------------------------------------------------
def show_distance_matrix(data, metric_name, p=None):
    if metric_name == 'minkowski':
        dist = pdist(data, metric=metric_name, p=p)
    else:
        dist = pdist(data, metric=metric_name)
    matrix = squareform(dist)
    print(f"\n--- {metric_name.capitalize()} Distance Matrix ---")
    print(np.round(matrix, 2))
    return matrix

euclid = show_distance_matrix(data, 'euclidean')
manhattan = show_distance_matrix(data, 'cityblock')
minkowski = show_distance_matrix(data, 'minkowski', p=3)

# --------------------------------------------------
# (c) Dendrograms for single, complete, and average linkage
# --------------------------------------------------
methods = ['single', 'complete', 'average']

plt.figure(figsize=(15, 5))
for i, method in enumerate(methods):
    plt.subplot(1, 3, i+1)
    linked = linkage(data, method=method, metric='euclidean')
    dendrogram(linked, labels=labels)
    plt.title(f'{method.capitalize()} Linkage')

plt.tight_layout()
plt.show()
```

```python
##Q1

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram

data = np.array([[18],[22],[25],[27],[42],[43]])
labels = ['18','22','25','27','42','43']

Z = linkage(data, method='single')
dendrogram(Z, labels=labels)
plt.title("Dendrogram - Single Linkage (Q1)")
plt.xlabel("Data points")
plt.ylabel("Distance")
plt.show()
```
